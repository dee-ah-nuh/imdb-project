```python
import pandas as pd
import numpy as np
```

```python
titleakas_url = 'https://datasets.imdbws.com/title.akas.tsv.gz'
titlebasics_url = 'https://datasets.imdbws.com/title.basics.tsv.gz'
ratings_url = 'https://datasets.imdbws.com/title.ratings.tsv.gz'
```

```python
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)
```

```python
ratings.replace({'\\N': np.nan})
```

```python
instructionsForRatings='Replace "/N" with np.nan (if any)'
ratings.to_csv('Data/ratings.csv')
```

```python
title = pd.read_csv(titlebasics_url, sep="\t", low_memory=False)
```

```python
title.replace({'\\N': np.nan})
```

```python
instructionsfortitle= '''
Replace '/N' with np.nan  
Eliminate movies that are null for runtimeMinutes 
Eliminate movies that are null for genre 
keep only titleType==Movie 
keep startYear 2000-2022 
Eliminate movies that include 'Documentary' in genre (see tip below)"
'''
title = title.dropna(subset=['genres'])
isDocumentary = title['genres'].str.contains('documentary', case=False)
title = title[~isDocumentary]
titleTypeMovie = title["titleType"]=='Movie'
title = title[titleTypeMovie]
title = title[title['genres'].dropna()]
title.to_csv("Data/title.csv")
```
